```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data.csv")
data = data.to_numpy()
x_train = data[:, 0]
y_train = data[:, 1]

plt.scatter(x_train, y_train)

plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
m = 0
c = 0
l = 0.0001
iterations = 50
n = float(len(x_train))
error = []
min_error=10**-5

for i in range(iterations):
    y_pred = m * x_train + c
    d1 = (-2 / n) * np.sum(x_train * (y_train - y_pred))
    d2 = (-2 / n) * np.sum(y_train - y_pred)

    error.append((1 / n) * np.sum(np.square(y_train - y_pred)))


    if i > 0 and np.abs(error[-1] - error[-2]) < min_error:
        break

    m = m - l * d1
    c = c - l * d2

print(f"Final Line : m={m} c={c}")
```

```python
plt.plot(error)
plt.xlabel("Iterations")
plt.ylabel("Error")
plt.show()
```

```python
y_pred = m * x_train + c

plt.scatter(x_train, y_train)
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(x_train, y_pred, color="red")
plt.show()
```
